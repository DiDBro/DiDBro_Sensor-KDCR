80个sensor， 10周测试，12failure, total accumulate down time is 96h, 

1)find MTBF, mfr
2)cal reliability over 6month
3) Three sensor are connected in a parallel arrangement ,the system work at least 2 sensor,calc reliability over 6 month

```python
import math

# ===== Given Data =====
num_sensors = 80
weeks = 10
failures = 12
downtime_hours = 96

# ===== 1) MTBF and Failure Rate =====
# Total calendar time (sensor-hours)
T_total = num_sensors * weeks * 7 * 24
# Total operating time = calendar time - downtime
T_op = T_total - downtime_hours

MTBF = T_op / failures  # hours
lambda_rate = 1 / MTBF   # failures per hour
MTTR = downtime_hours / failures  # hours

print("=" * 50)
print("1) MTBF and Failure Rate")
print("=" * 50)
print(f"Total calendar time: {T_total:,} sensor-hours")
print(f"Total operating time: {T_op:,} sensor-hours")
print(f"MTBF = {MTBF:,.2f} hours  ({MTBF/24:,.1f} days)")
print(f"Failure rate λ = {lambda_rate:.6e} failures/hour")
print(f"MTTR = {MTTR:.2f} hours")
print()

# ===== 2) Reliability over 6 Months =====
# 6 months = 6 * 30 * 24 = 4320 hours
t_6months = 6 * 30 * 24
R_single = math.exp(-lambda_rate * t_6months)

print("=" * 50)
print("2) Reliability over 6 Months (single sensor)")
print("=" * 50)
print(f"Time period: {t_6months} hours (6 × 30-day months)")
print(f"R(t) = exp(-λt) = exp(-{lambda_rate:.6e} × {t_6months})")
print(f"R(6 months) = {R_single:.4f}  ({R_single*100:.2f}%)")
print()

# ===== 3) 2-out-of-3 Parallel System =====
# R_sys = C(3,2)*R^2*(1-R) + C(3,3)*R^3 = 3R^2 - 2R^3
R_sys = 3 * R_single**2 - 2 * R_single**3

print("=" * 50)
print("3) 2-out-of-3 Parallel System Reliability")
print("=" * 50)
print(f"R_sys = 3R² - 2R³")
print(f"     = 3×({R_single:.4f})² - 2×({R_single:.4f})³")
print(f"     = 3×{R_single**2:.4f} - 2×{R_single**3:.4f}")
print(f"     = {3*R_single**2:.4f} - {2*R_single**3:.4f}")
print(f"R_sys(6 months) = {R_sys:.4f}  ({R_sys*100:.2f}%)")
print()

# ===== Summary Table =====
print("=" * 50)
print("SUMMARY TABLE")
print("=" * 50)
print(f"{'Metric':<35} {'Value':>15}")
print("-" * 50)
print(f"{'MTBF':<35} {MTBF:>15,.2f} hrs")
print(f"{'Failure rate λ':<35} {lambda_rate:>15.6e} /hr")
print(f"{'MTTR':<35} {MTTR:>15.2f} hrs")
print(f"{'Single sensor R(6mo)':<35} {R_single*100:>14.2f}%")
print(f"{'2-out-of-3 system R(6mo)':<35} {R_sys*100:>14.2f}%")
```
