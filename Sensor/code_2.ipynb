$$ RT = R_0 (1 + a T + b T^2)

T = 75

a = 3.85 * 10^-3

b = -5.8 * 10^{-7}

sigma_R0 = 0.08; sigma_a = 2.0 * 10^-5;sigma b = 1.6 * 10^-7

problem a) clculate mean resistance;

b) first order uncertainty propagation , calculate the standard deviation sigma_RT

c) approximate 95% interval for RT using Rt +- 2 sigma_RT

d) indentify which parameter cpntribute the largest variace to R_T at this tempurature

```python
import numpy as np

# Given parameters
T = 75
R0 = 100
a = 3.85e-3
b = -5.8e-7
sigma_R0 = 0.08
sigma_a = 2.0e-5
sigma_b = 1.5e-7

# -------------------------------------------------------------------
# (a) Mean resistance
# -------------------------------------------------------------------
# R_T = R0 * (1 + a*T + b*T^2)
RT_mean = R0 * (1 + a * T + b * T**2)
print(f"(a) Mean resistance R_T = {RT_mean:.6f} Ω")
print(f"    1 + aT + bT^2 = {1 + a*T + b*T**2:.8f}")

# -------------------------------------------------------------------
# (b) First-order uncertainty propagation
# -------------------------------------------------------------------
# Partial derivatives
dRT_dR0 = 1 + a * T + b * T**2      # ∂RT/∂R0
dRT_da  = R0 * T                      # ∂RT/∂a
dRT_db  = R0 * T**2                   # ∂RT/∂b

print(f"\nPartial derivatives:")
print(f"  ∂RT/∂R0 = {dRT_dR0:.6f}")
print(f"  ∂RT/∂a  = {dRT_da:.2f}")
print(f"  ∂RT/∂b  = {dRT_db:.2f}")

# Variance contributions (each term = (∂RT/∂x * σ_x)^2)
var_R0 = (dRT_dR0 * sigma_R0)**2
var_a  = (dRT_da  * sigma_a)**2
var_b  = (dRT_db  * sigma_b)**2

sigma_RT = np.sqrt(var_R0 + var_a + var_b)

print(f"\n(b) Variance contributions:")
print(f"  From R0: {var_R0:.8f}  ({var_R0/(var_R0+var_a+var_b)*100:.1f}%)")
print(f"  From a:  {var_a:.8f}  ({var_a/(var_R0+var_a+var_b)*100:.1f}%)")
print(f"  From b:  {var_b:.8f}  ({var_b/(var_R0+var_a+var_b)*100:.1f}%)")
print(f"\n  Total variance  σ²_RT = {var_R0 + var_a + var_b:.8f}")
print(f"  Standard dev    σ_RT  = {sigma_RT:.6f} Ω")

# -------------------------------------------------------------------
# (c) Approximate 95% confidence interval: RT ± 2·σ_RT
# -------------------------------------------------------------------
lower = RT_mean - 2 * sigma_RT
upper = RT_mean + 2 * sigma_RT
print(f"\n(c) Approximate 95% interval for R_T:")
print(f"  R_T ∈ [{lower:.4f}, {upper:.4f}] Ω")
print(f"  R_T = {RT_mean:.4f} ± {2*sigma_RT:.4f} Ω")

# -------------------------------------------------------------------
# (d) Which parameter contributes the largest variance?
# -------------------------------------------------------------------
contribs = {"R0": var_R0, "a": var_a, "b": var_b}
largest = max(contribs, key=contribs.get)
print(f"\n(d) Parameter '{largest}' contributes the largest variance "
      f"({contribs[largest]/(var_R0+var_a+var_b)*100:.1f}% of total).")
```
