2 identical 350 ohm metal foil strain gauges with G = 2.10, Gauge 1 is tensile Gauge 2 is compress, v_s = 3.30V. they are connect as the active arms of the half bridge.


a) for tensile strain magnitude xi = 600 * 10^-6, calc the R change of each active gauge

b) calc half bridge diff output magnitude

c) i f strain span of 0 to 1000* 10^6 occupy on 0,2.5V, output span ,whci amolifier gain is required

d) amplify signal is by 12-bit ADC with inputrange 0-3.30V what is strain resolution correspoinding to one ADC ?

e) if RMS noise at ADC is 2.00 mV,what is euivalents RMS strain nose afer avg 16 independent samples?

```python
import numpy as np

# ============================================================
# Given parameters
# ============================================================
R     = 350.0       # gauge resistance [ohm]
G     = 2.10        # gauge factor
Vs    = 3.30        # supply voltage [V]
eps   = 600e-6      # strain magnitude
eps_max = 1000e-6   # full-scale strain
V_out_span = 2.5    # desired output span [V]
ADC_bits  = 12      # 12-bit ADC
ADC_range = 3.30    # ADC input range [V]
V_noise_rms = 2.00e-3  # RMS noise at ADC [V]
N_avg    = 16       # number of independent samples averaged

# ============================================================
# (a) Resistance change of each active gauge
#     ΔR = G * ε * R
#     Gauge 1: tensile (+ε), Gauge 2: compressive (-ε)
# ============================================================
dR1 =  G *  eps * R      # Gauge 1 (tensile)
dR2 = -G *  eps * R      # Gauge 2 (compressive)

print("===== (a) Resistance change of each gauge =====")
print(f"ΔR1 (tensile)  = {dR1:.4f} Ω")
print(f"ΔR2 (compressive) = {dR2:.4f} Ω")
print()

# ============================================================
# (b) Half-bridge differential output magnitude
#     Vout = -Vs * G * ε / 2   (adjacent active arms)
# ============================================================
Vout = Vs * G * eps / 2          # magnitude
Vout_full = Vs * G * eps_max / 2 # full-scale output

print("===== (b) Half-bridge differential output =====")
print(f"|Vout| (at {eps*1e6:.0f} με) = {Vout*1e3:.4f} mV")
print(f"|Vout| (at {eps_max*1e6:.0f} με)  = {Vout_full*1e3:.4f} mV")
print()

# ============================================================
# (c) Amplifier gain required
#     Gain = V_out_span / Vout_full
# ============================================================
G_amp = V_out_span / Vout_full

print("===== (c) Amplifier gain =====")
print(f"Full-scale bridge output = {Vout_full*1e3:.4f} mV")
print(f"Desired output span       = {V_out_span:.2f} V")
print(f"Required gain G_amp       = {G_amp:.2f}")
print()

# ============================================================
# (d) Strain resolution (12-bit ADC, 0–3.30 V)
#     ADC LSB  = ADC_range / 2^bits
#     System sensitivity at ADC input = V_out_span / eps_max [V/ε]
#     Strain resolution = V_LSB / sensitivity
# ============================================================
V_LSB = ADC_range / (2**ADC_bits)
sensitivity = V_out_span / eps_max       # [V / ε]
eps_resolution = V_LSB / sensitivity

print("===== (d) Strain resolution =====")
print(f"ADC LSB voltage    = {V_LSB*1e3:.4f} mV")
print(f"System sensitivity = {sensitivity*1e6:.4f} V/με  (={V_out_span*1e3:.2f} mV / {eps_max*1e6:.0f} με)")
print(f"Strain resolution  = {eps_resolution*1e6:.4f} με per ADC count")
print()

# ============================================================
# (e) RMS strain noise after averaging 16 independent samples
#     V_noise_avg = V_noise_rms / sqrt(N)
#     ε_noise     = V_noise_avg / sensitivity
# ============================================================
V_noise_avg = V_noise_rms / np.sqrt(N_avg)
eps_noise = V_noise_avg / sensitivity

print("===== (e) RMS strain noise after averaging =====")
print(f"RMS noise (single sample) = {V_noise_rms*1e3:.2f} mV")
print(f"RMS noise after avg {N_avg}  = {V_noise_avg*1e3:.3f} mV")
print(f"Equivalent RMS strain noise = {eps_noise*1e6:.3f} με")
print()

# ============================================================
# Summary table
# ============================================================
print("=" * 50)
print("SUMMARY")
print("=" * 50)
print(f"ΔR per gauge                  : ±{dR1:.4f} Ω")
print(f"Bridge output (at 600 με)      : {Vout*1e3:.3f} mV")
print(f"Amplifier gain                : {G_amp:.1f}")
print(f"Strain resolution (1 ADC count): {eps_resolution*1e6:.2f} με")
print(f"RMS strain noise (16× avg)     : {eps_noise*1e6:.2f} με")
```
